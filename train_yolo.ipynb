```python
from ultralytics import YOLO
import os
```

```python
model = YOLO("models/yolov8n.pt")
```

```python
results = model.train(
    data="/Users/sabari/Documents/DS/GUVI/SmartVision-AI/smartvision_dataset/detection/data.yaml",
    epochs=15,
    imgsz=416,
    batch=4,
    project="/Users/sabari/Documents/DS/GUVI/SmartVision-AI/models",
    name="yolo_training"
)
```

```python
metrics = model.val()
print(metrics)
```
